### Preprocessing

```python
#Transform all the tweets to lower case
total_df['new_content'] = total_df['content'].apply(lambda x: " ".join(x.lower() for x in x.split()))
total_df['new_content'].head()
```

```python
#Remove all punctuations
total_df['new_content'] = total_df['new_content'].str.replace('[^\w\s]', '')
total_df['new_content'].head()
```

```python
#remove emojis if any
def remove_emoji(text):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags 
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)
total_df['new_content'] = total_df['new_content'].apply(lambda x: remove_emoji(x))
total_df['new_content'].head()
```

```python
#removing stop words
from nltk.tokenize import word_tokenize

# nltk.download('stopwords')
# nltk.download('punkt')
# Define stop words
stop_words = set(stopwords.words('english'))

# Function to remove stop words
def remove_stopwords(text):
    words = word_tokenize(text)
    filtered_sentence = [w for w in words if not w in stop_words]
    return " ".join(filtered_sentence)

total_df['n_content'] = total_df['new_content'].apply(remove_stopwords)
```

```python
n_df1 = total_df[['n_content', 'likeCount']]
```

```python
# Creating a lexicon of positive and negative words
positive_words = ["good", "great", "positive", "happy", "exciting", "satisfied", "beautiful", "majestic","radiant","thrilled","grateful","elated","blissful","pleased","delighted"]
negative_words = ["bad", "terrible", "negative", "sad", "boring", "terrorism", "lockdown",  "crime", "rob", "dirty", "gun", "thieves", "fight", "con","limited","expensive","unsafe" ]

# Function to classify tweets based on the lexicon
def classify_tweet(text):
    positive_count = 0
    negative_count = 0
    words = text.split()
    for word in words:
        if word in positive_words:
            positive_count += 1
        elif word in negative_words:
            negative_count += 1
    if positive_count > negative_count:
        return "positive"
    elif negative_count > positive_count:
        return "negative"
    else:
        return "neutral"

n_df1['sentiment'] = n_df1['n_content'].apply(classify_tweet)

n_df1.head()
```
